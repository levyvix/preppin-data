```python
def tweak_df(df):
    return df.assign(
        **{
            "Bank": lambda df_: df_["Transaction Code"].str.split("-").str[0],
            "Online or In-Person": lambda df_: df_["Online or In-Person"].map(
                {1: "Online", 2: "In-Person"}
            ),
            "Transaction Date": lambda df_: pd.to_datetime(
                df_["Transaction Date"], dayfirst=True
            ).dt.day_of_week.map(
                {
                    0: "1 - Monday",
                    1: "2 - Tuesday",
                    2: "3 - Wednesday",
                    3: "4 - Thursday",
                    4: "5 - Friday",
                    5: "6 - Saturday",
                    6: "7 - Sunday",
                }
            ),
        }
    )
```

```python
df_clean = tweak_df(df)
```
